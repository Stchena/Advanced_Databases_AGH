```python
# Ex10
def heat(m, v):
    return m*v**2 / 2
def kmph_to_mps(v):
    return v/3.6

m = 2500 #[g]
v = kmph_to_mps(60) #[km/h to m/s]
Q = heat(m,v)
print("Heat in Joules:%.f" % Q)
print("Heat in Kcals:%.f" % (Q/4186.8))

m = 3000
v = np.linspace(kmph_to_mps(200),0)
v_log = np.linspace(kmph_to_mps(200),0)
Q = heat(m,v)
Q_log = heat(m,v_log)
plt.plot(v,Q)
plt.show()
plt.semilogy(v_log,Q_log)
plt.xlim(0,kmph_to_mps(200))
plt.show()
```
